```python
#@title OLS
X = sm.add_constant(X)
model = sm.OLS(y, X)
result = model.fit()
# T√≠nh to√°n residual (th√†nh ph·∫ßn ng·∫´u nhi√™n)
residuals = result.resid

# Ki·ªÉm tra t·ª± t∆∞∆°ng quan trong th√†nh ph·∫ßn ng·∫´u nhi√™n
autocorrelation = residuals.autocorr()

print("T·ª± t∆∞∆°ng quan: ", autocorrelation)
print(result.summary())
```

```python
#@title FEM
# Run PanelOLS on the structured panel data
X_fem = sm.add_constant(X)
model = PanelOLS(y, X_fem, entity_effects=True)
results = model.fit()
print(results)
```

```python
#@title REM
# Run Random Effects Model (REM)
X_rem = sm.add_constant(X)
rem_model = RandomEffects(y, X_rem)
rem_results = rem_model.fit()
print(rem_results)
```

```python
#@title D-I-D
# Create interaction term between treatment and post
df['treatment_post'] = df['treat'] *  df['post']

# Define the dependent variable (y) and independent variables (X)
y = df[['roa']]
X = df[['treat', 'post', 'treatment_post','esgscore',  'logassets','logmkcap', 'capint', 'revenuegrowth', 'debttotalassets', 'currentratio']]
X = sm.add_constant(X)
# Run the Difference-in-Differences model
model = sm.OLS(y, X)
result = model.fit()
print(result.summary())
```
